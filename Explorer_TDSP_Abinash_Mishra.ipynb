**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Crashes_20250106.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x= average_crashes_per_hour.index, y= average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby(data['CRASH DATE'].dt.to_period("D")).size()
daily_crashes.index = daily_crashes.index.to_timestamp()
# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Crashes', fontsize=14)
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

````python
#Exploring features related to weather
unique_conditions = data['Weather_Condition'].unique().compute()
unique_conditions
````

````python
import numpy as np
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd

# Extract unique weather categories
weather_categories = data['Weather_Condition'].dropna().unique().compute()

# Use TF-IDF to vectorize weather categories
vectorizer = TfidfVectorizer()
weather_vectors = vectorizer.fit_transform(weather_categories)

# Apply KMeans clustering
num_clusters = 10  # Define the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
labels = kmeans.fit_predict(weather_vectors)

# Map weather categories to their respective clusters
cluster_mapping = pd.DataFrame({
    'Weather_Condition': weather_categories,
    'Cluster_Label': labels
}).sort_values(by='Cluster_Label')

# Assign clusters to the main dataset
cluster_dict = dict(zip(weather_categories, labels))
data['Weather_Cluster'] = data['Weather_Condition'].map(cluster_dict)

# Display the cluster mapping
print("Weather Condition to Cluster Mapping:")
cluster_mapping

# Save the cluster mapping for reference
cluster_mapping.to_csv('weather_cluster_mapping.csv', index=False)
print("\nCluster mapping saved to 'weather_cluster_mapping.csv'")

# Inspect each cluster's content
print("\nWeather Conditions in Each Cluster:")
for cluster in range(num_clusters):
    print(f"Cluster {cluster}:")
    print(cluster_mapping[cluster_mapping['Cluster_Label'] == cluster]['Weather_Condition'].values)
    print("\n")


````

````python
#Representing clusters as groups

cluster_sizes = cluster_mapping['Cluster_Label'].value_counts().sort_index()  # Cluster sizes
num_clusters = len(cluster_sizes)

# Group weather conditions by cluster
clusters_content = cluster_mapping.groupby('Cluster_Label')['Weather_Condition'].apply(list)

# Circle properties
circle_radius = 20 # Scale radius based on cluster size
colors = plt.cm.viridis(np.linspace(0, 1, num_clusters))  # Generate colors for clusters

# Create figure
fig, ax = plt.subplots(figsize=(14, 10))

# Set axis limits
ax.set_xlim(-1.5 * num_clusters, 1.5 * num_clusters)
ax.set_ylim(-1.5 * num_clusters, 1.5 * num_clusters)

# Plot circles for each cluster
angles = np.linspace(0, 2 * np.pi, num_clusters, endpoint=False)  # Arrange circles in a circular layout
for i, (label, content) in enumerate(clusters_content.items()):
    x = np.cos(angles[i]) * num_clusters  # X-coordinate
    y = np.sin(angles[i]) * num_clusters  # Y-coordinate

    # Draw the circle
    circle = plt.Circle((x, y), radius=circle_radius * 0.2, color=colors[i], alpha=0.6)
    ax.add_artist(circle)

    # Add the label and contents inside the circle
    ax.text(
        x, y,
        f"Cluster {label}\n" + "\n".join(content[:5]) + ("\n..." if len(content) > 5 else ""),
        ha='center', va='center', fontsize=8, color='white', wrap=True
    )

# Final touches
ax.set_title("Weather Clusters with Contents", fontsize=16)
ax.set_aspect('equal')  # Ensure circles are perfectly round
ax.axis('off')  # Turn off axes
plt.show()
````

````python
# Analyze the contents of each cluster and mapping them
# interpretation
clusters_content = cluster_mapping.groupby('Cluster_Label')['Weather_Condition'].apply(list)

# Assign descriptive names to each cluster
cluster_names = {
    0: "Clear/Dusty Weather",
    1: "Windy Conditions",
    2: "Rainy Weather",
    3: "Snowy Conditions",
    4: "Thunderstorms",
    5: "Cloudy Weather",
    6: "Snow/Rain Showers",
    7: "Hazy/Smoky Conditions",
    8: "Storm Conditions",
    9: "Heavy or Light Drizzle"
}

# Map cluster names to the clusters
cluster_mapping['Cluster_Name'] = cluster_mapping['Cluster_Label'].map(cluster_names)
data['Weather_Cluster_Name'] = data['Weather_Cluster'].map(cluster_names)
````

````python
# Group by clusters and severity, and compute counts
severity_by_cluster = data.groupby(['Weather_Cluster_Name', 'Severity']).size().compute()

# Reset the index
severity_by_cluster = severity_by_cluster.reset_index()

# Rename columns for clarity
severity_by_cluster.columns = ['Weather_Cluster_Name', 'Severity', 'Count']

# Pivot manually using Pandas after computing
severity_pivot = severity_by_cluster.pivot_table(index='Weather_Cluster_Name', columns='Severity', values='Count', fill_value=0)

# Plot the severity distribution across weather clusters
severity_pivot.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title("Severity Distribution Across Weather Clusters", fontsize=16)
plt.xlabel("Weather Cluster", fontsize=14)
plt.ylabel("Number of Accidents", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Severity", fontsize=12)
plt.tight_layout()
plt.show()
````

````python
#Removing crictical missing values to reduce computational cost
data = data.dropna(subset=['Wind_Speed(mph)', 'Temperature(F)', 'Visibility(mi)', 'Weather_Condition'])
````

````python
#Building LightGBM Classifier Model to predict and see the dependency of weather conditions on the severity of accidents
data['Weather_Cluster_Name'] = data['Weather_Cluster_Name'].fillna('Unknown')
data['Weather_Cluster_Name'] = data['Weather_Cluster_Name'].astype(str)

# Define features
features = ['Temperature(F)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Weather_Cluster_Name']

# Define target
X = data[features]
y = data['Severity']
````

````python
from sklearn.model_selection import train_test_split
X_pd = X.compute()
y_pd = y.compute()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pd, y_pd, test_size=0.2, random_state=42, stratify=y_pd)
````

````python
import lightgbm as lgb

# Convert categorical columns to 'category' type
X_train['Weather_Cluster_Name'] = X_train['Weather_Cluster_Name'].astype('category')
X_test['Weather_Cluster_Name'] = X_test['Weather_Cluster_Name'].astype('category')

# Initialize and train the LightGBM model
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(X_train, y_train, categorical_feature=['Weather_Cluster_Name'])
````

````python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Make predictions
y_pred = lgb_model.predict(X_test)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
````

Obtained an accuracy of 85%
